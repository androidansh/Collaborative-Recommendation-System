```python
import pandas as pd
```

```python
ratings = pd.read_csv("dataset/ratings.csv")
movies = pd.read_csv("dataset/movies.csv")
```

```python
finalDf = ratings.merge(movies, on='movieId')
```

```python
finalDf.drop(['timestamp','genres'],inplace=True,axis=1)
finalDf.head()
```

```python
finalDf.to_csv('final_df.csv', index=False)
```

```python
dataset = ratings.merge(movies, on='movieId')
```

```python
dataset['genres'] = (
    dataset['genres']
    .dropna()                    # handle NaN
    .str.replace('|', ' ', regex=False)
    .str.lower()                   # normalize (optional but recommended)
    .str.strip()
)
```

```python
dataset.drop(['timestamp'],inplace=True,axis=1)
```

```python
dataset.to_csv("dataset.csv",index=False)
```
